```python
import pandas as pd
import plotly.graph_objects as go
import datetime as dt
```

```python
p = "EUR_USD"
gran = "H4"
df = pd.read_pickle(f"../data/{p}_{gran}.pkl")
```

```python
df
```

```python
df['sTime'] = [dt.datetime.strftime(x, "%y-%m-%d %H:%M") for x in df.time]
df
```

```python
print(df[:100].sTime)
fig = go.Figure()
fig.add_trace(go.Candlestick(
        x = df[:100].sTime,
        open=df[:100].mid_o,
        low=df[:100].mid_l,
        high=df[:100].mid_h,
        close=df[:100].mid_c,
    )    
)
fig.update_layout(xaxis = dict(type="category"))
fig.update_xaxes(
    nticks=5
)
fig.show()
```
